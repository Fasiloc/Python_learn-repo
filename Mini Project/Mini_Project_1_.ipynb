```python
import random
q1='''Q.Which character is used in Python to make a single line comment?
a. /
b. //
c. #
d. !''' 

q2='''Q.Which of the following statements is correct regarding the object-oriented programming concept in Python?
a. Classes are real-world entities while objects are not real
b. Objects are real-world entities while classes are not real
c. Both objects and classes are real-world entities
d. All of the above'''

q3='''Q.What is the method inside the class in python language?
a. Object
b. Function
c. Attribute
d. Argument'''

q4='''Q.Which of the following operators is the correct option for power(ab)?
a. a ^ b
b. a**b
c. a ^ ^ b
d. a ^ * b'''

q5='''Q.Which of the following functions is a built-in function in python language?
a. val()
b. print()
c. Print()
d. None of these'''

q6='''Q.) How many control statements python supports?
a. Four
b. Five
c. Three
d. None of the these'''

q7='''Q.How many keywords present in the python programming language?
a. 32
b. 61
c. 33
d. 27'''


quiz={q1:'c',q2:'b',q3:'b',q4:'b',q5:'b',q6:'c',q7:'c'}
users=['User 1','User 2','User 3','User 4','User 5']
quizd=[]
for i in users:
  mark=0
  i=input('Enter your name : ')
  for j in range(len(quiz)):
    q,a=random.choice(list(quiz.items()))
    quizd.append(q)
    print('\n',q,'\n')
    c=input('Enter your choice a,b,c,d. \n For skip this question enter "s". \n For terminate quiz enter "t" \n\n Choice :')
    if q in quizd:
      continue
    elif c=='s':
      continue
    elif c=='t':
      break
    elif c==a:
      print('\nCorrect answer\n')
      mark+=3  
    else:
      print('\noops wrong answer\n')
      mark-=1
 print(i,'Quiz Finished. Your total mark is :',mark,'\n')
```
